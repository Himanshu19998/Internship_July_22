# Relation between Gender and Specialization
Both are categorical columns, using chi square testing we can find relation between them.

```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
pd.crosstab(df.Specialization,df.Gender,margins=True)
```

```python
observed = pd.crosstab(df.Specialization,df.Gender)
observed
```

```python
chi2_contingency(observed)
```

```python
#computing chi2 test statistic,p-value,degree of freedom

chi2_test_stat = chi2_contingency(observed)[0]
p_val = chi2_contingency(observed)[1]
dof = chi2_contingency(observed)[2]
                                   
```

```python
confidence_level = 0.95
alpha=1-confidence_level
chi2_critical = chi2.ppf(1-alpha, dof)
chi2_critical
```

```python
#plotting the chi2 distribution
x_min = 0
x_max = 100
x= np.linspace(x_min, x_max, 100)
y = chi2.pdf(x, dof)
plt.xlim(x_min,x_max)
plt.plot(x,y)
chi2_critical_right = chi2_critical
x1= np.linspace(chi2_critical_right,x_max,100)
y1= chi2.pdf(x1,dof)
plt.fill_between(x1, y1, color='Blue')
```

```python
if(chi2_test_stat> chi2_critical):
  print('Reject Null Hypothesis')
else:
  print('Fail to Reject Null Hypothesis')
    

```

```python
if(p_val< alpha):
  print('Reject Null Hypothesis')
else:
  print('Fail to Reject Null Hypothesis')
    
```

# Observation
For 95% Confidence, these two categorical columns are highly correlated.
